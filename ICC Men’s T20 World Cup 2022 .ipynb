# T20 World Cup 2022 Analysis using Python



**Now let’s start this task by importing the necessary Python libraries and the dataset: **

```python

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data = pd.read_csv("t20-world-cup-22.csv")
print(data.head())
```

The special command `%matplotlib inline` to ensure that our plots are shown and embedded within the Jupyter notebook itself. Without this command, sometimes plots may show up in pop-up windows.

### Now let’s look at the number of matches won by each team in the world cup:

```python
winner = data["winner"].value_counts()
label = winner.index
counts = winner.values
plt.figure(figsize=(15, 7))
plt.title("Number of Matches Won by teams in t20 World Cup 2022")
plt.xlabel('Winner')
plt.ylabel('Count')
plt.bar(label,counts);
```

As England won the t20 world cup 2022, England won five matches. And Both Pakistan and India won 4 matches.


**Now let’s have a look at the number of matches won by batting first or second in the t20 world cup 2022:**

```python
def absolute_value(val):
    a  = int(np.round(val/100.*values.sum()))
    return a

won_by = data["won by"].value_counts()

label = won_by.index
values = won_by.values
total = values.sum()

colors = ['gold','lightgreen']
plt.figure(figsize=(15, 7))
plt.title("Number of Matches Won Runs or Wickets")
plt.pie(values,labels = label,colors = colors,autopct=absolute_value,startangle=90) # autopct is used to display the values in number format
plt.show();
```

So in the t20 world cup 2022, 16 matches were won by batting first, and 13 matches were won by chasing. Now, let’s have a look at the toss decisions by teams in the world cup:

```python
def absolute_value(val):
    a  = int(np.round(val/100.*counts.sum()))
    return a

toss = data["toss decision"].value_counts()
print(toss)
label = toss.index
counts = toss.values
colors = ['skyblue','yellow']
plt.figure(figsize=(15, 7))
plt.title("Toss Decisions in t20 World Cup 2022")
plt.pie(counts,labels = label,colors = colors,autopct=absolute_value,startangle=90) # autopct is used to display the values in number format
plt.show();
```

So in 17 matches, the teams decided to bat first, and in 13 matches, the teams chose to field first. Now let’s have a look at the top scorers in the t20 world cup 2022:

```python
top_score = data.dropna().groupby(['top scorer'])['highest score'].sum()
plt.figure(figsize=(20, 10))
plt.title("Top Scorers in t20 World Cup 2022", fontsize=40)
plt.xlabel('Top Scorer',fontsize=30)
plt.ylabel('Highest Score', fontsize=30)
ax = sns.barplot(x=top_score.index, y=top_score.values,
                 data=data,
                 errwidth=0)

plt.rc('xtick', labelsize=20, ) 
plt.rc('ytick', labelsize=20, ) 
plt.xticks(rotation='vertical')
ax.bar_label(ax.containers[0]);
```


So, Virat Kohli scored the highest 208 runs. Undoubtedly, he was the best batsman in the t20 world cup 2022. Now let’s have a look at the number of player of the match awards in the world cup:

```python
man_of_the_match = data.dropna().groupby(['player of the match'])['player of the match'].count()
plt.figure(figsize=(15, 7))
sns.set_style("whitegrid")
plt.title("Player of the Match Awards in T20 World Cup 2022", fontsize=20)
plt.xlabel('Player of the Match',fontsize=20)
plt.ylabel('Counts', fontsize=20)
ax = sns.barplot(x=man_of_the_match.index, y=man_of_the_match.values,
                 data=data,
                 errwidth=0)

plt.rc('xtick', labelsize=10, ) 
plt.rc('ytick', labelsize=10, ) 
plt.ylim(0,3)
plt.xticks(rotation='vertical')
ax.bar_label(ax.containers[0]);
```

Virat Kohli, Sam Curran, Taskin Ahmed, Suryakumar Yadav, and Shadab Khan got the player of the match in 2 matches. No player got the player of the match award in more than two matches.

Now let’s have a look at the bowlers with the best bowling figures at the end of the matches:

```python
bowler_of_the_match = data['best bowler'].value_counts()
plt.figure(figsize=(15, 7))
sns.set_style("whitegrid")
plt.title("Best Bowlers in t20 World Cup 2022", fontsize=20)
plt.xlabel('Player of the Match',fontsize=20)
plt.ylabel('Counts', fontsize=20)
ax = sns.barplot(x=bowler_of_the_match.index, y=bowler_of_the_match.values,
                 data=data,
                 errwidth=0, color='b')

plt.rc('xtick', labelsize=10, ) 
plt.rc('ytick', labelsize=10, ) 
plt.ylim(0,4)
plt.xticks(rotation='vertical')
ax.bar_label(ax.containers[0]);
```

Sam Curran was the only best bowler in 3 matches. Undoubtedly, he deserved to be the player of the tournament. Now let’s compare the runs scored in the first innings and second innings in every stadium of the t20 world cup 2022:

```python
s1 = data.dropna().groupby('venue')["first innings score",'second innings score'].sum()
s2 = data.dropna().groupby('venue')["second innings score"].sum()
plt.figure(figsize=(15, 7))
sns.set_style("whitegrid")
plt.title("Best Stadiums to Bat First or Chase", fontsize=20)
width = 0.25
bar1=plt.bar(s1.index,s1['first innings score'].values,width,color='r',alpha=0.8);
bar2 = plt.bar(s1.index,s1['second innings score'].values,2*width, color='b',alpha=0.4);
plt.legend( (bar1, bar2), ('First Inning', 'Second Inning') )
plt.show()
```

So SCG was the only stadium in the world cup that was best for batting first. Other stadiums didn’t make much difference while batting first or chasing.

Now let’s compare the number of wickets lost in the first innings and second innings in every stadium of the t20 world cup 2022:

```python
s1 = data.dropna().groupby('venue')["first innings wickets",'second innings wickets'].sum()
plt.figure(figsize=(15, 7))
sns.set_style("whitegrid")
plt.title("Best Statiums to Bowl First or Defend", fontsize=20)
width = 0.25
bar1=plt.bar(s1.index,s1['first innings wickets'].values,width,color='g',alpha=0.8);
bar2 = plt.bar(s1.index,s1['second innings wickets'].values,2*width, color='y',alpha=0.4);
plt.legend( (bar1, bar2), ('First Inning', 'Second Inning') )
plt.show()
```

SCG was the best stadium to bowl while defending the target. While the Optus Stadium was the best stadium to bowl first.
